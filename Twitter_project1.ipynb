```python
import pandas as pd
import snscrape.modules.twitter as sntwitter

# Define search query and maximum number of tweets to scrape
search_query = "Elon Musk"
max_tweets = 1000

# Create empty list to hold scraped tweet data
tweets_list = []

# Loop through tweets using snscrape library
for i, tweet in enumerate(sntwitter.TwitterSearchScraper(search_query).get_items()):
    if i >= max_tweets:
        break
    
    # Extract relevant tweet data
    date = tweet.date.strftime('%Y-%m-%d %H:%M:%S')
    tweet_id = tweet.id
    url = tweet.url
    content = tweet.content
    user = tweet.user.username
    reply_count = tweet.replyCount
    retweet_count = tweet.retweetCount
    language = tweet.lang
    source = tweet.sourceLabel
    like_count = tweet.likeCount
    
    # Append tweet data to list
    tweets_list.append([date, tweet_id, url, content, user, reply_count, retweet_count, language, source, like_count])

# Create dataframe from scraped tweet data
tweets_df = pd.DataFrame(tweets_list, columns=['Date', 'Tweet ID', 'URL', 'Content', 'User', 'Reply Count', 'Retweet Count', 'Language', 'Source', 'Like Count'])

# Display dataframe
print(tweets_df.head())
```

```python
import pymongo
import snscrape.modules.twitter as sntwitter
from datetime import datetime, timedelta

# Define search query and maximum number of tweets to scrape
search_query = "Elon Musk"
max_tweets = 1000

# Create empty list to hold scraped tweet data
tweets_list = []

# Loop through tweets using snscrape library
for i, tweet in enumerate(sntwitter.TwitterSearchScraper(search_query + ' since:' + (datetime.today() - timedelta(days=100)).strftime('%Y-%m-%d')).get_items()):
    if i >= max_tweets:
        break
    
    # Extract relevant tweet data
    date = tweet.date.strftime('%Y-%m-%d %H:%M:%S')
    tweet_id = tweet.id
    url = tweet.url
    content = tweet.content
    user = tweet.user.username
    reply_count = tweet.replyCount
    retweet_count = tweet.retweetCount
    language = tweet.lang
    source = tweet.sourceLabel
    like_count = tweet.likeCount
    
    # Append tweet data to list
    tweets_list.append({'date': date, 'tweet_id': tweet_id, 'url': url, 'content': content, 'user': user, 'reply_count': reply_count, 'retweet_count': retweet_count, 'language': language, 'source': source, 'like_count': like_count})

# Connect to MongoDB
client = pymongo.MongoClient('mongodb+srv://Hari:H1234567@cluster0.3y1yrb1.mongodb.net/')
db = client['twitter']
collection = db['tweets']

# Store scraped tweet data in MongoDB collection
collection.insert_one({'Scraped Word': search_query, 'Scraped Date': datetime.today().strftime('%Y-%m-%d'), 'Scraped Data': tweets_list})

# Confirm that data has been stored in MongoDB
print(collection.find_one({'Scraped Word': search_query}))
```
